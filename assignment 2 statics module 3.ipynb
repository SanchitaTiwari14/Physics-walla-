```python
#1. What is hypothesis testing in statistics

#Hypothesis testing in statistics is a method used to determine whether there is enough statistical 
#evidence in favor of a specific belief or hypothesis about a population parameter. It involves making an 
#assumption (called a hypothesis), collecting sample data, and then using statistical tests to assess whether the
#data supports or contradicts the hypothesis.
#Null and Alternative Hypotheses:
#Null Hypothesis (H‚ÇÄ): The assumption that there is no effect or no difference in the population.
#Alternative Hypothesis (H‚ÇÅ or Ha): The assumption that there is an effect or difference.
```

```python
#2. What is the null hypothesis, and how does it differ from the alternative hypothesis

# the null hypothesis represents the status quo or no effect scenario, while the alternative hypothesis 
# proposes that a significant effect or relationship exists. Hypothesis testing seeks to provide evidence
# to either reject or fail to reject the null hypothesis in favor of the alternative hypothesis.

# Key Differences:

# Null Hypothesis (H‚ÇÄ): Assumes no effect, no difference, or no relationship.
# Alternative Hypothesis (H‚ÇÅ or Ha): Assumes that there is an effect, difference, or relationship.


# Null Hypothesis (H‚ÇÄ): We try to reject the null hypothesis based on evidence from the sample.
# Alternative Hypothesis (H‚ÇÅ or Ha): We accept or support the alternative hypothesis if the null hypothesis is rejected.
```

```python
#3. What is the significance level in hypothesis testing, and why is it important

# The significance level is a crucial aspect of hypothesis testing that defines the threshold for deciding
# whether the evidence is strong enough to reject the null hypothesis. It helps to control the probability of making Type I 
#     errors and ensures that conclusions drawn from statistical tests are reliable and meaningful.
```

```python
#4. What does a P-value represent in hypothesis testing
# The P-value is a key concept in hypothesis testing. It helps to determine whether the
# observed data is consistent with the null hypothesis or provides strong enough evidence to reject it.
# A small p-value (usually ‚â§ 0.05) indicates strong evidence against the null hypothesis, suggesting that the 
# null hypothesis is unlikely and may be rejected in favor of the alternative hypothesis.
# A large p-value (> 0.05) indicates weak evidence against the null hypothesis, suggesting that there is not 
# enough evidence to reject the null hypothesis.
```

```python
#5 How do you interpret the P-value in hypothesis testing

# If p-value ‚â§ Œ± (usually 0.05): This means the result is statistically significant. You can reject the null hypothesis and say there‚Äôs enough evidence to support the alternative hypothesis.
# If p-value > Œ± (usually 0.05): This means the result is not statistically significant. You fail to reject the null hypothesis because there‚Äôs not enough evidence to support the alternative hypothesis.
```

```python
#6. What are Type 1 and Type 2 errors in hypothesis testing

# In hypothesis testing, there are two types of errors that can occur when making a 
# decision about the null hypothesis (H‚ÇÄ):

# Type 1 Error: You say something is true (reject the null) when it‚Äôs not.
# Type 2 Error: You miss something that‚Äôs true (fail to reject the null when it‚Äôs false)
```

```python
#7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing



# In hypothesis testing, the choice between a one-tailed or two-tailed test depends on 
# the direction of the effect you're testing for.

#  A one-tailed test is used when the research hypothesis (alternative hypothesis) is testing 
# for an effect in one specific direction (either greater than or less than a certain value).

# A two-tailed test is used when the research hypothesis (alternative hypothesis) is testing for 
# an effect in both directions‚Äîwhether the sample statistic is either greater than or less than a certain value.
```

```python
#10. @What is the T-distribution, and when should it be used instead of the normal distribution

# The T-distribution (also called Student's t-distribution) is a type of probability distribution that 
# is symmetric and bell-shaped, like the normal distribution, but with heavier tails. This means it gives 
# more probability to extreme values compared to the normal distribution. The T-distribution is used in
# hypothesis testing and confidence intervals when the sample size is small and/or the population 
# standard deviation is unknown.


# When to Use the Normal Distribution Instead of the T-distribution:
# You can use the normal distribution (Z-test) in the following cases:

# Sample Size is Large (n > 30):

# When the sample size is large, the Central Limit Theorem suggests that the sampling distribution of the sample mean will approach a normal distribution, even if the underlying data is not normal. In this case, the T-distribution and the normal distribution become very similar, so the normal distribution can be used.
# Population Standard Deviation is Known:

# If the population standard deviation (œÉ) is known (e.g., from previous studies or other data), then you can use the Z-test and the normal distribution. This is because the normal distribution does not require estimation of the standard deviation from the sample.
```

```python
#20 @ What is an ANOVA test, and what are its assumptions


# ANOVA (Analysis of Variance) is a statistical test used to determine whether there are any statistically significant differences between the means of three or more independent groups. Instead of comparing all possible pairs of group means (which would be inefficient), ANOVA tests the overall variation between group means and compares it to the variation within the groups.

# The main goal of ANOVA is to test the null hypothesis that all group means are equal against the alternative hypothesis that at least one group mean is different.

# For ANOVA to provide valid results, several assumptions must be met:

# Independence:
# The observations in each group should be independent of each other. That is, the outcome for one observation should not influence the outcome of another observation.
# Normality:
# The data in each group should be approximately normally distributed. If the data is heavily skewed or has outliers, the results may not be reliable.
# Homogeneity of Variances (Homoscadasticity):
# The variances in each group should be approximately equal. This assumption can be tested using the Levene's test or Bartlett's test.
```

```python
#1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

import math
import scipy.stats as stats

# Function to perform Z-test
def z_test(sample_mean, population_mean, sample_std, sample_size, significance_level=0.05):
    # Calculate Z-score
    z_score = (sample_mean - population_mean) / (sample_std / math.sqrt(sample_size))
    
    # Find the p-value for the Z-score
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test
    
    # Print the Z-score and p-value
    print(f"Z-score: {z_score}")
    print(f"P-value: {p_value}")
    
    # Interpretation
    if p_value < significance_level:
        print("Reject the null hypothesis: There is a significant difference between the sample mean and population mean.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and population mean.")

# Example usage:
sample_mean = 105  # Example sample mean
population_mean = 100  # Known population mean
sample_std = 15  # Sample standard deviation
sample_size = 50  # Sample size

# Perform the Z-test
z_test(sample_mean, population_mean, sample_std, sample_size)
```

```python
#3 Implement a one-sample Z-test using Python to compare the sample mean with the population mean@
import numpy as np
import scipy.stats as stats

# Function to perform the one-sample Z-test
def one_sample_z_test(sample_data, population_mean, population_std, significance_level=0.05):
    # Step 1: Calculate the sample mean
    sample_mean = np.mean(sample_data)
    
    # Step 2: Calculate the Z-score
    sample_size = len(sample_data)
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    
    # Step 3: Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test
    
    # Step 4: Output the Z-score and p-value
    print(f"Sample Mean: {sample_mean}")
    print(f"Z-score: {z_score}")
    print(f"P-value: {p_value}")
    
    # Step 5: Interpretation
    if p_value < significance_level:
        print("Reject the null hypothesis: There is a significant difference between the sample mean and population mean.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and population mean.")

# Simulate some sample data (for example, sample size = 30, sample data is normally distributed)
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=105, scale=15, size=30)  # Sample mean = 105, population std = 15, sample size = 30

# Population parameters
population_mean = 100  # Known population mean
population_std = 15    # Known population standard deviation

# Perform the one-sample Z-test
one_sample_z_test(sample_data, population_mean, population_std)
```

```python
#4 Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Function to perform the two-tailed Z-test
def two_tailed_z_test(sample_data, population_mean, population_std, significance_level=0.05):
    # Step 1: Calculate the sample mean
    sample_mean = np.mean(sample_data)
    
    # Step 2: Calculate the Z-score
    sample_size = len(sample_data)
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    
    # Step 3: Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test
    
    # Step 4: Output the Z-score and p-value
    print(f"Sample Mean: {sample_mean}")
    print(f"Z-score: {z_score}")
    print(f"P-value: {p_value}")
    
    # Step 5: Interpretation
    if p_value < significance_level:
        print("Reject the null hypothesis: There is a significant difference between the sample mean and population mean.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and population mean.")
    
    return z_score, p_value

# Simulate some sample data (for example, sample size = 30, sample data is normally distributed)
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=105, scale=15, size=30)  # Sample mean = 105, population std = 15, sample size = 30

# Population parameters
population_mean = 100  # Known population mean
population_std = 15    # Known population standard deviation

# Perform the two-tailed Z-test
z_score, p_value = two_tailed_z_test(sample_data, population_mean, population_std)

# Step 6: Visualize the decision region

# Define the x values for the plot (range of Z-scores)
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)  # Probability density function of the standard normal distribution

# Plot the standard normal curve
plt.plot(x, y, label="Standard Normal Distribution", color="blue")

# Plot the rejection regions (Critical Z-values)
alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha / 2)  # The critical Z-value for a two-tailed test
plt.fill_between(x, y, where=(x <= -critical_value), color="red", alpha=0.5, label="Rejection Region (Left Tail)")
plt.fill_between(x, y, where=(x >= critical_value), color="red", alpha=0.5, label="Rejection Region (Right Tail)")

# Mark the Z-score on the plot
plt.axvline(z_score, color="green", linestyle="--", label=f"Z-score: {z_score:.2f}")
plt.axvline(-z_score, color="green", linestyle="--")

# Add labels and title
plt.title("Two-Tailed Z-Test and Decision Region")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
#5 @Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Function to calculate and visualize Type I and Type II errors
def type_1_and_type_2_errors(population_mean, sample_mean, population_std, sample_size, significance_level=0.05):
    # Step 1: Define parameters
    # Population mean (H0 mean) and standard deviation
    critical_value = stats.norm.ppf(1 - significance_level)  # Critical Z-value for significance level
    beta_value = stats.norm.cdf(critical_value - (sample_mean - population_mean) / (population_std / np.sqrt(sample_size)))

    # Step 2: Type I Error (alpha): Probability of rejecting the null hypothesis when it's true
    # Type I error is the area under the curve to the right of the critical value
    alpha = 1 - stats.norm.cdf(critical_value)
    
    # Step 3: Type II Error (beta): Probability of failing to reject the null hypothesis when the alternative is true
    # Type II error is the area under the alternative hypothesis curve (shifted by the difference between sample mean and population mean)
    # This is the area under the curve to the left of the critical value
    beta = 1 - beta_value

    # Step 4: Print results
    print(f"Type I Error (alpha): {alpha:.4f}")
    print(f"Type II Error (beta): {beta:.4f}")

    # Step 5: Visualization
    x = np.linspace(-4, 4, 1000)  # X range for plotting
    y_null = stats.norm.pdf(x, 0, 1)  # Null hypothesis (H0) distribution (standard normal)
    
    # Shifted distribution for alternative hypothesis (H1)
    shift = (sample_mean - population_mean) / population_std
    y_alt = stats.norm.pdf(x, shift, 1)  # Alternative hypothesis distribution (shifted normal)

    plt.figure(figsize=(10, 6))

    # Plot null hypothesis (H0) and alternative hypothesis (H1)
    plt.plot(x, y_null, label="Null Hypothesis (H0)", color="blue")
    plt.plot(x, y_alt, label="Alternative Hypothesis (H1)", color="green")
    
    # Fill rejection regions (Type I Error) on the right tail
    plt.fill_between(x, 0, y_null, where=(x >= critical_value), color="red", alpha=0.5, label="Rejection Region (Type I Error)")
    
    # Fill the Type II error region under H1 curve
    plt.fill_between(x, 0, y_alt, where=(x <= critical_value), color="yellow", alpha=0.5, label="Type II Error Region")

    # Mark critical value for rejection region
    plt.axvline(x=critical_value, color="black", linestyle="--", label=f"Critical Value: {critical_value:.2f}")

    # Add labels and title
    plt.title(f"Type I and Type II Errors in Hypothesis Testing\nŒ± = {significance_level}, Œ≤ = {beta:.4f}")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()

    plt.grid(True)
    plt.show()

# Example usage:
population_mean = 0   # Null hypothesis mean (H0)
sample_mean = 1       # Alternative hypothesis mean (H1)
population_std = 1    # Population standard deviation
sample_size = 30      # Sample size
significance_level = 0.05  # Significance level for the test

# Perform the function to calculate and visualize Type I and Type II errors
type_1_and_type_2_errors(population_mean, sample_mean, population_std, sample_size, significance_level)
```

```python
#6 Write a Python program to perform an independent T-test and interpret the results
import numpy as np
from scipy import stats

# Function to perform the independent T-test
def independent_t_test(group1_data, group2_data, significance_level=0.05):
    # Step 1: Perform the independent T-test
    t_stat, p_value = stats.ttest_ind(group1_data, group2_data)

    # Step 2: Print the results
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    
    # Step 3: Interpretation of the result
    if p_value < significance_level:
        print("Reject the null hypothesis: There is a significant difference between the two groups.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference between the two groups.")

# Example data for two independent groups (e.g., group 1 and group 2)
np.random.seed(42)  # For reproducibility

# Simulate data for two groups
group1 = np.random.normal(loc=100, scale=15, size=30)  # Group 1: mean = 100, std = 15, sample size = 30
group2 = np.random.normal(loc=110, scale=20, size=30)  # Group 2: mean = 110, std = 20, sample size = 30

# Perform the independent T-test
independent_t_test(group1, group2)
```

```python
#7 Perform a paired sample T-test using Python and visualize the comparison results
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Function to perform the paired sample T-test
def paired_t_test(group1_data, group2_data, significance_level=0.05):
    # Step 1: Perform the paired sample T-test
    t_stat, p_value = stats.ttest_rel(group1_data, group2_data)
    
    # Step 2: Print the results
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    
    # Step 3: Interpretation of the result
    if p_value < significance_level:
        print("Reject the null hypothesis: There is a significant difference between the two related groups.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference between the two related groups.")

# Example data for two related groups (e.g., pre-treatment and post-treatment)
np.random.seed(42)  # For reproducibility

# Simulate data for two related groups
group1 = np.random.normal(loc=100, scale=15, size=30)  # Group 1: pre-treatment data
group2 = group1 + np.random.normal(loc=5, scale=10, size=30)  # Group 2: post-treatment data (shifted by 5 units)

# Perform the paired sample T-test
paired_t_test(group1, group2)

# Visualization of the comparison results
plt.figure(figsize=(10, 6))

# Plotting the data as a scatter plot to show the change from group1 to group2
plt.scatter(group1, group2, color='blue', label="Data points", alpha=0.7)

# Add line connecting points to emphasize the difference
for i in range(len(group1)):
    plt.plot([group1[i], group2[i]], [group1[i], group2[i]], color='gray', alpha=0.5)

# Set labels and title
plt.xlabel("Pre-Treatment Data (Group 1)")
plt.ylabel("Post-Treatment Data (Group 2)")
plt.title("Comparison of Pre-Treatment and Post-Treatment Data (Paired Sample T-Test)")
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```

```python
#8 Simulate data and perform both Z-test and T-test, then compare the results using Python

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Function to perform the Z-test
def z_test(sample_data, population_mean, population_std, sample_size):
    sample_mean = np.mean(sample_data)
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed test
    return z_stat, p_value

# Function to perform the T-test
def t_test(sample_data, population_mean, sample_size):
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation
    t_stat, p_value = stats.ttest_1samp(sample_data, population_mean)
    return t_stat, p_value

# Simulate data for the sample group (for example, a group of 30 samples)
np.random.seed(42)  # For reproducibility
sample_size = 30
population_mean = 100
population_std = 15
sample_data = np.random.normal(loc=102, scale=15, size=sample_size)  # Sample mean is 102, same population std as assumed

# Perform Z-test and T-test
z_stat, z_p_value = z_test(sample_data, population_mean, population_std, sample_size)
t_stat, t_p_value = t_test(sample_data, population_mean, sample_size)

# Print results
print("Z-test Results:")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {z_p_value:.4f}")

print("\nT-test Results:")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {t_p_value:.4f}")

# Comparison and interpretation
if z_p_value < 0.05:
    print("\nZ-test: Reject the null hypothesis - There is a significant difference from the population mean.")
else:
    print("\nZ-test: Fail to reject the null hypothesis - There is no significant difference from the population mean.")

if t_p_value < 0.05:
    print("\nT-test: Reject the null hypothesis - There is a significant difference from the population mean.")
else:
    print("\nT-test: Fail to reject the null hypothesis - There is no significant difference from the population mean.")

# Visualizing the sample data distribution and population mean
plt.figure(figsize=(10, 6))
plt.hist(sample_data, bins=10, alpha=0.7, label='Sample Data')
plt.axvline(population_mean, color='red', linestyle='--', label=f'Population Mean ({population_mean})')
plt.title('Sample Data Distribution and Population Mean')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
from scipy import stats

def confidence_interval(sample_data, confidence_level=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    - sample_data (array-like): The sample data.
    - confidence_level (float): The confidence level (default is 0.95 for 95% confidence).
    
    Returns:
    - lower_bound (float): The lower bound of the confidence interval.
    - upper_bound (float): The upper bound of the confidence interval.
    """
    # Sample size
    n = len(sample_data)
    
    # Sample mean
    sample_mean = np.mean(sample_data)
    
    # Sample standard deviation
    sample_std = np.std(sample_data, ddof=1)  # Use ddof=1 for sample std deviation
    
    # T-score (since population standard deviation is unknown)
    t_score = stats.t.ppf((1 + confidence_level) / 2, df=n-1)
    
    # Margin of error
    margin_of_error = t_score * (sample_std / np.sqrt(n))
    
    # Confidence Interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return lower_bound, upper_bound

# Example usage
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Simulate sample data

# Calculate the confidence interval for the sample mean
lower_bound, upper_bound = confidence_interval(sample_data, confidence_level=0.95)

print(f"95% Confidence Interval for the sample mean: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
#10 Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
from scipy import stats

def margin_of_error(sample_data, confidence_level=0.95):
    """
    Calculate the margin of error for a sample mean at a given confidence level.

    Parameters:
    - sample_data (array-like): The sample data.
    - confidence_level (float): The confidence level (default is 0.95 for 95% confidence).
    
    Returns:
    - margin_of_error (float): The margin of error.
    """
    # Sample size
    n = len(sample_data)
    
    # Sample standard deviation
    sample_std = np.std(sample_data, ddof=1)  # Use ddof=1 for sample std deviation
    
    # T-score (since population standard deviation is unknown)
    t_score = stats.t.ppf((1 + confidence_level) / 2, df=n-1)
    
    # Margin of error
    margin_of_error = t_score * (sample_std / np.sqrt(n))
    
    return margin_of_error

# Example usage
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Simulate sample data

# Calculate the margin of error for the sample mean
moe = margin_of_error(sample_data, confidence_level=0.95)

print(f"Margin of Error at 95% confidence level: {moe:.2f}")
```

```python
#11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayesian_inference(prior, likelihood, evidence):
    """
    Perform Bayesian Inference using Bayes' Theorem.

    Parameters:
    - prior (float): The prior probability of the hypothesis (P(H)).
    - likelihood (float): The likelihood of the evidence given the hypothesis (P(E|H)).
    - evidence (float): The probability of observing the evidence (P(E)).

    Returns:
    - posterior (float): The posterior probability (P(H|E)).
    """
    posterior = (likelihood * prior) / evidence
    return posterior

# Example values:
# Prior probability (P(H)) - Probability that a person has the disease (e.g., 1% of the population)
prior = 0.01

# Likelihood (P(E|H)) - Probability that a person tests positive given they have the disease (sensitivity of the test)
likelihood = 0.9

# Evidence (P(E)) - Probability of testing positive regardless of whether the person has the disease.
# This can be calculated as the probability of testing positive both for diseased and non-diseased.
p_e_given_h = likelihood  # Likelihood for diseased
p_e_given_not_h = 0.05   # False positive rate (5% chance of testing positive when not having the disease)

# Total probability of evidence P(E) = P(E|H)P(H) + P(E|not H)P(not H)
evidence = (p_e_given_h * prior) + (p_e_given_not_h * (1 - prior))

# Perform Bayesian inference to find the posterior probability (P(H|E))
posterior = bayesian_inference(prior, likelihood, evidence)

print(f"The posterior probability (P(H|E)) that a person has the disease given a positive test result: {posterior:.4f}")
```

```python
#13  Write a Python program to calculate the expected frequencies for a Chi-square test based on observed dataD
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate the expected frequencies for a Chi-square test.

    Parameters:
    - observed (ndarray): 2D array or matrix representing the observed frequencies.

    Returns:
    - expected (ndarray): 2D array of expected frequencies.
    """
    # Calculate the row totals and column totals
    row_totals = np.sum(observed, axis=1, keepdims=True)
    column_totals = np.sum(observed, axis=0, keepdims=True)
    
    # Calculate the total sum of all observed frequencies
    total = np.sum(observed)
    
    # Calculate the expected frequencies using the formula
    expected = (row_totals @ column_totals) / total
    
    return expected

# Example: Observed data for a contingency table
observed_data = np.array([[30, 5],  # 30 males prefer coffee, 5 males prefer tea
                          [10, 15]]) # 10 females prefer coffee, 15 females prefer tea

# Calculate the expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

# Print the observed and expected frequencies
print("Observed Frequencies:")
print(observed_data)

print("\nExpected Frequencies:")
print(expected_frequencies)
```

```python
#16 Implement an F-test using Python to compare the variances of two random samples
import numpy as np
from scipy.stats import f

# Function to perform an F-test
def f_test(sample1, sample2):
    """
    Perform an F-test to compare the variances of two samples.

    Parameters:
    - sample1: First sample data (list or array).
    - sample2: Second sample data (list or array).
    
    Returns:
    - f_statistic: The calculated F-statistic.
    - p_value: The p-value for the test.
    """
    # Calculate variances for both samples
    var1 = np.var(sample1, ddof=1)  # Sample variance for sample1
    var2 = np.var(sample2, ddof=1)  # Sample variance for sample2
    
    # The F-statistic is the ratio of the larger variance to the smaller variance
    f_statistic = var1 / var2 if var1 > var2 else var2 / var1
    
    # Degrees of freedom
    df1 = len(sample1) - 1  # Degrees of freedom for sample1
    df2 = len(sample2) - 1  # Degrees of freedom for sample2
    
    # Calculate the p-value for the F-test using the cumulative distribution function (CDF)
    p_value = 1 - f.cdf(f_statistic, df1, df2)  # For two-tailed test
    
    return f_statistic, p_value

# Simulate two random samples from normal distributions
np.random.seed(42)  # For reproducibility
sample1 = np.random.normal(loc=50, scale=10, size=100)  # Sample 1 with mean=50, std=10
sample2 = np.random.normal(loc=50, scale=20, size=100)  # Sample 2 with mean=50, std=20

# Perform the F-test
f_statistic, p_value = f_test(sample1, sample2)

# Print the results
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject the null hypothesis: The variances are significantly different.")
else:
    print("\nFail to reject the null hypothesis: The variances are not significantly different.")
```

```python
#19  Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVAD

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(groups):
    """
    Check the assumptions of normality, independence, and equal variance for ANOVA.
    
    Parameters:
    - groups: List of arrays, where each array represents one group of data.
    
    Returns:
    - None. Prints results of normality, independence, and equal variance tests.
    """
    
    # 1. Check Normality assumption using Shapiro-Wilk test
    print("Checking normality assumption using Shapiro-Wilk test:")
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i + 1}: Shapiro-Wilk test statistic = {stat:.3f}, p-value = {p_value:.3f}")
        if p_value < 0.05:
            print(f"  - Group {i + 1} is NOT normally distributed (reject H0).")
        else:
            print(f"  - Group {i + 1} is normally distributed (fail to reject H0).")
    
    # 2. Check Equal Variance assumption using Levene's test
    print("\nChecking equal variance assumption using Levene's test:")
    stat, p_value = stats.levene(*groups)
    print(f"Levene's test statistic = {stat:.3f}, p-value = {p_value:.3f}")
    if p_value < 0.05:
        print("  - Variances are significantly different (reject H0).")
    else:
        print("  - Variances are equal (fail to reject H0).")

    # 3. Check Independence assumption
    # Assuming independence based on the design of the experiment, so no statistical test for this.
    print("\nChecking independence assumption:")
    print("  - Assuming data samples are independent based on experimental design.")
    
    # Visualize the distribution of the groups to assess normality and variance visually
    plt.figure(figsize=(12, 6))
    
    # Plotting a boxplot to visually check variance and outliers
    sns.boxplot(data=groups)
    plt.title("Boxplot of the Groups")
    plt.xlabel("Groups")
    plt.ylabel("Values")
    plt.show()
    
    # Plotting a QQ plot for each group to check normality
    for i, group in enumerate(groups):
        plt.figure(figsize=(6, 6))
        stats.probplot(group, dist="norm", plot=plt)
        plt.title(f"QQ Plot for Group {i + 1}")
        plt.show()

# Example usage with simulated data
np.random.seed(42)

# Simulate data for 3 groups with different means
group1 = np.random.normal(loc=50, scale=10, size=100)
group2 = np.random.normal(loc=55, scale=10, size=100)
group3 = np.random.normal(loc=60, scale=10, size=100)

# Check assumptions for ANOVA
check_anova_assumptions([group1, group2, group3])
```

```python
#20 Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Simulate data for two factors: 'Factor1' and 'Factor2'
np.random.seed(42)

# Factor 1 (e.g., two levels: "A", "B")
factor1 = np.random.choice(['A', 'B'], size=200)

# Factor 2 (e.g., two levels: "X", "Y")
factor2 = np.random.choice(['X', 'Y'], size=200)

# Response variable (dependent variable)
# We simulate data for each combination of Factor1 and Factor2, with different means
response = np.random.normal(0, 5, 200)  # base noise
response[factor1 == 'A'][factor2 == 'X'] += 5  # Effect of A-X
response[factor1 == 'B'][factor2 == 'Y'] += 7  # Effect of B-Y
response[factor1 == 'A'][factor2 == 'Y'] += 3  # Effect of A-Y

# Create a DataFrame
data = pd.DataFrame({
    'Factor1': factor1,
    'Factor2': factor2,
    'Response': response
})

# Perform Two-Way ANOVA with interaction using statsmodels
model = ols('Response ~ C(Factor1) * C(Factor2)', data=data).fit()
anova_table = anova_lm(model)

# Print the ANOVA table
print(anova_table)

# Interpretation of results
alpha = 0.05
if anova_table['PR(>F)'][0] < alpha:
    print("\nReject the null hypothesis: Factor1 has a significant effect.")
else:
    print("\nFail to reject the null hypothesis: Factor1 does not have a significant effect.")

if anova_table['PR(>F)'][1] < alpha:
    print("Reject the null hypothesis: Factor2 has a significant effect.")
else:
    print("Fail to reject the null hypothesis: Factor2 does not have a significant effect.")

if anova_table['PR(>F)'][2] < alpha:
    print("Reject the null hypothesis: There is a significant interaction between Factor1 and Factor2.")
else:
    print("Fail to reject the null hypothesis: There is no significant interaction between Factor1 and Factor2.")

# Visualize the interaction between Factor1 and Factor2 using an interaction plot
plt.figure(figsize=(8, 6))
sns.pointplot(x='Factor1', y='Response', hue='Factor2', data=data, dodge=True, markers='o', linestyles='-', ci=None)
plt.title("Interaction Plot for Factor1 and Factor2", fontsize=15)
plt.xlabel("Factor1", fontsize=12)
plt.ylabel("Response", fontsize=12)
plt.show()
```

```python
#21  Write a Python program to visualize the F-distribution and discuss its use in hypothesis testingD


import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def plot_f_distribution(dfn_values, dfd_values, x_range=(0, 5), num_points=1000):
    """
    Visualizes the F-distribution for different degrees of freedom for the numerator and denominator.
    
    Parameters:
    - dfn_values: List of numerator degrees of freedom (dfn) values.
    - dfd_values: List of denominator degrees of freedom (dfd) values.
    - x_range: The range of values for the x-axis (F-statistic values).
    - num_points: Number of points in the x range for plotting.
    """
    x = np.linspace(x_range[0], x_range[1], num_points)

    plt.figure(figsize=(10, 6))
    
    # Plot the F-distribution for different dfn and dfd values
    for dfn in dfn_values:
        for dfd in dfd_values:
            # Calculate the F-distribution PDF (probability density function)
            y = stats.f.pdf(x, dfn, dfd)
            plt.plot(x, y, label=f'dfn={dfn}, dfd={dfd}')
    
    # Add labels and title
    plt.title("F-distribution for Different Degrees of Freedom", fontsize=14)
    plt.xlabel("F-statistic", fontsize=12)
    plt.ylabel("Probability Density", fontsize=12)
    plt.legend(title="dfn and dfd", loc="upper right")
    plt.grid(True)
    plt.show()

# Example usage: Visualize the F-distribution with different degrees of freedom
dfn_values = [1, 2, 5]  # Different numerator degrees of freedom (dfn)
dfd_values = [1, 2, 5]  # Different denominator degrees of freedom (dfd)

plot_f_distribution(dfn_values, dfd_values)
```

```python
#22  Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Simulate data for 3 groups with different means
np.random.seed(42)

# Generate random data for three groups
group1 = np.random.normal(50, 10, 100)  # Mean=50, Std=10, N=100
group2 = np.random.normal(55, 12, 100)  # Mean=55, Std=12, N=100
group3 = np.random.normal(60, 15, 100)  # Mean=60, Std=15, N=100

# Combine the data into a DataFrame
data = pd.DataFrame({
    'value': np.concatenate([group1, group2, group3]),
    'group': ['Group 1'] * 100 + ['Group 2'] * 100 + ['Group 3'] * 100
})

# Perform One-Way ANOVA to test if there's a significant difference between the groups
f_statistic, p_value = stats.f_oneway(group1, group2, group3)

# Print the results
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# Interpretation of the results
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: At least one group mean is significantly different.")
else:
    print("\nFail to reject the null hypothesis: No significant difference between the group means.")

# Visualize the results with a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='group', y='value', data=data, palette="Set2")
plt.title("One-Way ANOVA: Group Mean Comparison", fontsize=15)
plt.xlabel("Group", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.show()
```

```python
#23 Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means


import numpy as np
import scipy.stats as stats

# Step 1: Simulate random data from a normal distribution
np.random.seed(42)  # For reproducibility
sample_size = 100
population_mean = 50
population_std = 10

# Simulate data for the sample
sample_data = np.random.normal(population_mean, population_std, sample_size)

# Step 2: Set up the hypothesis
# Null hypothesis (H‚ÇÄ): The mean of the sample is equal to the population mean (Œº = 50)
# Alternative hypothesis (H‚ÇÅ): The mean of the sample is not equal to the population mean (Œº ‚â† 50)

# Step 3: Perform a one-sample t-test
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation
t_statistic, p_value = stats.ttest_1samp(sample_data, population_mean)

# Step 4: Interpretation of the results
alpha = 0.05  # Significance level

print(f"Sample Mean: {sample_mean}")
print(f"Sample Standard Deviation: {sample_std}")
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("\nReject the null hypothesis: There is a significant difference between the sample mean and the population mean.")
else:
    print("\nFail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")
```

```python
#24  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the resultsD

import numpy as np
from scipy import stats

# Step 1: Simulate random data (sample data)
np.random.seed(42)  # For reproducibility
sample_size = 50
population_variance = 25  # Hypothesized population variance

# Simulating data from a normal distribution
sample_data = np.random.normal(0, np.sqrt(population_variance), sample_size)

# Step 2: Calculate the sample variance
sample_variance = np.var(sample_data, ddof=1)  # Sample variance (ddof=1 for unbiased estimate)

# Step 3: Perform the Chi-square test for population variance
# Chi-square test statistic: (n - 1) * sample_variance / population_variance
test_statistic = (sample_size - 1) * sample_variance / population_variance

# Degrees of freedom = n - 1
df = sample_size - 1

# Calculate the p-value (two-tailed test)
p_value = 2 * min(stats.chi2.cdf(test_statistic, df), 1 - stats.chi2.cdf(test_statistic, df))

# Step 4: Interpret the results
alpha = 0.05  # Significance level

print(f"Sample Variance: {sample_variance}")
print(f"Test Statistic (Chi-square): {test_statistic}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("\nReject the null hypothesis: The population variance is significantly different from the hypothesized value.")
else:
    print("\nFail to reject the null hypothesis: There is no significant difference between the population variance and the hypothesized value.")
```

```python
#25 Write a Python script to perform a Z-test for comparing proportions between two datasets or groupsD
import numpy as np
import scipy.stats as stats

def z_test_proportions(success_1, size_1, success_2, size_2):
    # Step 1: Calculate sample proportions
    p1 = success_1 / size_1
    p2 = success_2 / size_2

    # Step 2: Calculate the pooled sample proportion
    p = (success_1 + success_2) / (size_1 + size_2)

    # Step 3: Calculate the standard error
    SE = np.sqrt(p * (1 - p) * (1 / size_1 + 1 / size_2))

    # Step 4: Calculate the Z statistic
    Z = (p1 - p2) / SE

    # Step 5: Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(Z)))

    # Step 6: Return Z statistic and p-value
    return Z, p_value

# Example usage:
# Sample data: 
# Group 1: 50 successes out of 200
# Group 2: 60 successes out of 250
success_1 = 50
size_1 = 200
success_2 = 60
size_2 = 250

# Perform the Z-test for proportions
Z_stat, p_value = z_test_proportions(success_1, size_1, success_2, size_2)

# Output the results
print(f"Z-statistic: {Z_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: The proportions are significantly different.")
else:
    print("\nFail to reject the null hypothesis: There is no significant difference between the proportions.")

```

```python
#26 D Implement an F-test for comparing the variances of two datasets, then interpret and visualize the resultsD

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def f_test_variances(data1, data2):
    # Step 1: Calculate the sample variances
    var1 = np.var(data1, ddof=1)  # Sample variance for group 1 (ddof=1 for unbiased estimate)
    var2 = np.var(data2, ddof=1)  # Sample variance for group 2

    # Step 2: Calculate the F-statistic (larger variance / smaller variance)
    F_stat = var1 / var2 if var1 >= var2 else var2 / var1

    # Step 3: Calculate the degrees of freedom
    df1 = len(data1) - 1  # Degrees of freedom for group 1
    df2 = len(data2) - 1  # Degrees of freedom for group 2

    # Step 4: Calculate the p-value from the F-distribution
    p_value = 2 * min(stats.f.cdf(F_stat, df1, df2), 1 - stats.f.cdf(F_stat, df1, df2))

    # Step 5: Return the F-statistic and p-value
    return F_stat, p_value

# Example usage
# Simulated data for two groups
np.random.seed(42)
data1 = np.random.normal(0, 10, 100)  # Group 1 (mean=0, std=10)
data2 = np.random.normal(0, 20, 100)  # Group 2 (mean=0, std=20)

# Perform F-test to compare variances
F_stat, p_value = f_test_variances(data1, data2)

# Output the results
print(f"F-statistic: {F_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: The variances are significantly different.")
else:
    print("\nFail to reject the null hypothesis: The variances are not significantly different.")

# Visualize the results using boxplots to compare the distributions of the two datasets
plt.figure(figsize=(8, 6))
plt.boxplot([data1, data2], labels=['Group 1', 'Group 2'])
plt.title("Comparison of Group 1 and Group 2 Variances")
plt.ylabel("Values")
plt.show()
```

```python
#27   Perform a Chi-square test for goodness of fit with simulated data and analyze the results

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulate observed data (e.g., frequency of categories in a dataset)
np.random.seed(42)
observed = np.random.choice([1, 2, 3, 4, 5], size=500, p=[0.1, 0.2, 0.3, 0.25, 0.15])

# Calculate observed frequencies
observed_counts = np.array([np.sum(observed == i) for i in range(1, 6)])

# Set expected probabilities (e.g., uniform distribution, all categories are equally likely)
expected_probabilities = np.array([1/5] * 5)  # Uniform distribution
expected_counts = expected_probabilities * len(observed)

# Perform the Chi-square goodness of fit test
chi2_stat, p_value = stats.chisquare(observed_counts, expected_counts)

# Output the results
print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: The observed frequencies do not match the expected frequencies.")
else:
    print("\nFail to reject the null hypothesis: The observed frequencies match the expected frequencies.")

# Visualize the observed and expected frequencies
categories = ['1', '2', '3', '4', '5']

plt.figure(figsize=(8, 6))
plt.bar(categories, observed_counts, alpha=0.6, label="Observed", width=0.4, align="center")
plt.bar(categories, expected_counts, alpha=0.6, label="Expected", width=0.4, align="edge")
plt.xlabel("Categories")
plt.ylabel("Frequency")
plt.title("Chi-square Goodness of Fit Test")
plt.legend()
plt.show()
```
